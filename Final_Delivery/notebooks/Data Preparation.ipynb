# Relational Data Model- defining a primary key

In the relational model of databases, a primary key is a specific choice of a minimal set of attributes (columns) that uniquely specify a tuple (row) in a relation (table) 

A primary key’s main features are:

1.It must contain a unique value for each row of data.<br>
2.It cannot contain null values.

```python
data_path='../data/raw/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
pd_raw=pd.read_csv(data_path)
pd_raw.head()
```

**NOTE:**<br>
Latitude(Lat) and Longitude(Long): Master data (do not change) #remove

The Dates: Transaction data(do change)

```python
pd_data_base = pd_raw.rename(columns ={'Country/Region':'country',
                                      'Province/State':'state'})


pd_data_base = pd_data_base.drop(['Lat','Long'],axis=1)
pd_data_base.head()
```

**Now construction of Primary Key with date, state and country**

```python
pd_data_base.set_index(['state','country']) #set state and country as index
```

```python
#pd_data_base.set_index(['state','country']).T #transposed inoder to have dates in time dimension
```

```python
test_pd = pd_data_base.set_index(['state','country']).T
```

```python
#test_pd.columns
```

```python
test_pd.stack(level=[0,1]).reset_index() #to bring multidimension columns to row based
```

```python
pd_test_model=test_pd.rename(columns={'level_0':'date',
                         0:'confirmed'})
```

```python
pd_test_model
```

```python
pd_relational_model=pd_data_base.set_index(['state','country']) \
                                .T                              \
                                .stack(level=[0,1])             \
                                .reset_index()                  \
                                .rename(columns={'level_0':'date',
                                                0:'confirmed'})
pd_relational_model.head()
```

```python
pd_relational_model.dtypes
```

```python
pd_relational_model['date'] = pd_relational_model.date.astype('datetime64[ns]') #convert date which is till a str type to datetime type
```

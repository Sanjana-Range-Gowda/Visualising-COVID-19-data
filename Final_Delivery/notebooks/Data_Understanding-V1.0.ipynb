**Web Scrapping**

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.rki.de/DE/Content/InfAZ/N/Neuartiges_Coronavirus/Fallzahlen.html')
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup.get_text() #full web content
```

```python
html_table = soup.find('table') #need to find the names in Developer tool
```

```python
all_rows = html_table.find_all('tr')
```

```python
final_data_list=[]
```

```python
#sweeping through all rows and cleaning the data table
for pos,rows in enumerate(all_rows):
    #print(pos)
    #print(rows)
    
    #for each_col in rows.find_all('td'):
        #print(each_col.get_text(strip=True))
    
    col_list=[each_col.get_text(strip=True) for each_col in rows.find_all('td')] #wrapping to function so that its more stable
    #print(col_list)
    final_data_list.append(col_list)
    
```

```python
pd_daily_status=pd.DataFrame(final_data_list).dropna().rename(columns={0:'state',
                                                      1:'cases',
                                                      2:'difference',
                                                      3:'last_7days',
                                                      4:'incidence',
                                                      5:'deaths'})
```
